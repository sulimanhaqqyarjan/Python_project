### Scrapped Data from 2022-11-01 to 2023-07-10

```python
start_page = 1
end_page = 5
```

```python
titles = []
links = []
web_links = []
dates = []
news = []
```

```python
i=1
while start_page <= end_page:
    url = 'https://khybernews.tv/pu/قامى-خبرونه/page/{}'.format(start_page)

    html = requests.get(url).content
    sel = Selector(text = html)

    links = sel.xpath('//h3[@class="entry-title td-module-title"]/a/@href').extract()

    for link in links:
        print(link, i)
        html = requests.get(link).content
        sel = Selector(text = html)

        titles.append(sel.xpath('//h1[@class="entry-title"]/text()').extract_first())
        dates.append(sel.xpath('//time[@class="entry-date updated td-module-date"]/text()').extract_first())
        news.append(sel.xpath('//div[@class="td-post-content tagdiv-type"]//text()').extract())
        web_links.append(sel.xpath('//link[@rel="canonical"]/@href').extract_first())

        if(i%15==0):
            time.sleep(30)

        i+=1

    if(i%4==0):
        time.sleep(4)

    start_page += 1
```

### Save Data into Pandas Dataframe

```python
df = pd.DataFrame(columns=['Headline', 'Published Date', 'Web Link', 'News'])

df.Headline = titles
df['Published Date'] = dates
df['Web Link'] = web_links
df.News = news
```
