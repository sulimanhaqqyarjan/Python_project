```python
import spacy

# Load the spaCy language model
nlp = spacy.load("en_core_web_sm")

# Process a sample text
text = "100 له پاکستان وروسته ایران د افغانستان هغه ګاونډی دی چې ډېر افغان کډوال پکې مېشت دي او دې هېواد هم له ډېرې مودې راهیسې د دوی په ټکو د  افغان کډوالو د نیولو او $1 بېرته ۵  ستنولو لړۍ جاري ساتلې"
doc = nlp(text)

# Print token information (token and token_name)
for token in doc:
    print(token.text, token.pos_)  # token.text for the token itself, token.pos_ for part of speech
```

```python
import pandas as pd
import spacy

# Load the spaCy language model
nlp = spacy.load('en_core_web_sm')

# Sample DataFrame with a 'News' column
data = {'News': ['News']}
df = pd.DataFrame(data)

# Create a list to store tokenized data
tokenized_data = []

# Process and tokenize each text entry in the 'News' column
for text_entry in df['News']:
    doc = nlp(text_entry)
    tokens = [token.text for token in doc]
    tokenized_data.append(tokens)

# Add the tokenized data as a new column in the DataFrame
df['Tokens'] = tokenized_data

# Display the DataFrame with the tokenized data
print(df)
```
